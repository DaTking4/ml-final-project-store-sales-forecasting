## Exploratory Data Analysis

### Load and Sort Features Data

```python
import pandas as pd
import matplotlib.pyplot as plt

features = pd.read_csv('data/features.csv', parse_dates=['Date'])
features = features.sort_values(['Store', 'Date']).reset_index(drop=True)

MARKDOWNS = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
CONTINUOUS = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment']

features.head()
```

### Missing Value Analysis

```python
null_pct = features[CONTINUOUS + MARKDOWNS].isnull().mean().sort_values(ascending=False) * 100

fig, ax = plt.subplots(figsize=(10, 4))
null_pct.plot(kind='bar', ax=ax)
ax.set_ylabel('Missing (%)')
ax.set_title('Missing values per feature')
plt.tight_layout()
plt.show()
```
